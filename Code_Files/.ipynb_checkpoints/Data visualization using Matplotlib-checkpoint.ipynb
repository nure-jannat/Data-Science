```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ls = [7, 9, 17, 45, 48, 51, 52, 63, 65, 72, 130]
df = pd.DataFrame({"ls" : ls})
Q1 = df['ls'].quantile(0.25)
Q3 = df['ls'].quantile(0.75)

IQR = Q3 -Q1
print(f'Q1 : {Q1}, Q3 : {Q3}, IQR : {IQR}')
print('lower limit : ', Q1 - (1.5*IQR))
print('upper limit : ', Q3 + (1.5*IQR))
```

```python
data = np.array(ls)
fig = plt.figure(figsize=(10,7))
plt.boxplot(data)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
x = np.array(x)
y = x ** 2
y1 = x ** 3
plt.plot(x, y, "r--")
```

```python
fig1 = plt.figure(figsize=(10,7))

axes = fig1.add_axes([0, 0, 0.5, 0.5])
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")
axes.plot(x, y, label= "square curve")
axes.plot(x, y1, label= "cube curve")
axes.legend(loc = )
```
